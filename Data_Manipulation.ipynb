```python
import pandas as pd
import random

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('poker-hand-training-true.data', sep = ",")
```

```python
def count_hands():
    arr = [];
    for i in range(10):
        arr.append(0)
    
    last_column = df[df.columns[10]]
    for i in range(len(last_column)):
        arr[last_column[i]] += 1 
    
    return arr
```

```python
def dupliacation_prob():
    hand_counts = count_hands()
    dupliacation_prob = []
    
    for i in range(len(hand_counts)):
        dupliacation_prob.append( (hand_counts[0] - hand_counts[i]) / hand_counts[i])
    
    return dupliacation_prob
```

```python
def update_df(df):
    duplicate_prob = dupliacation_prob()
    first_size = len(df)
    
    for i in range(first_size):
        current_row = df.iloc[i]
        dup_times = duplicate_prob[current_row[10]]
        
        while(dup_times >= 1):
            df = df.append(current_row)
            dup_times -= 1
        rnd = random.random()
        if(dup_times > rnd):
            df = df.append(current_row)
    return df
```

```python
df = update_df(df)
```
